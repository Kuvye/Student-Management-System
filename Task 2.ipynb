 # STUDENT RESULT MANAGEMENT SYSTEM

 ## Project Description: This project is designed to help Bright Future Academy manage student records efficiently.

 ### Features
Be able to:
- Add student records
- View all students
- Update student scores
- Search for students
- Automatically calculate grades
- Generate class statistics


### Tools Used
- Python Dictionary

```python
#Dictionary to store student names and scores

Students = {}
```

## Grade Calculation Function



```python
#This function accepts a student's score and returns the appropriate grade based on the grading system.

def CalculateGrade(Score):
    if 70 <= Score <=100:
        return "A"
    elif 60 <= Score <=69:
        return "B"
    elif 50 <= Score <=59:
        return "C"
    elif 45 <= Score <=49:
        return "D"
    else:
        return "F"
```

## Add Student Function
 This function collects student information and scores it in the dictionary

```python
#to create add student function

def AddStudent ():
    name = input("Enter Student name: ")
    try:
        Score = float(input("Enter Score: "))
        
        if Score < 0 or Score > 100:
            print("Score must be between 0 and 100")
            return

        Students[name] = Score
        print("Student successfully added")

    except ValueError:
        print("Invalid Format/Score Entered")
```

## View Student Records

This function displays all students along sides their scores and grades 

```python
def ViewStudents ():
    if not Students:
        print("No Records Found!!")
        return
        
    print("Name\tScore\tGrade")

    for name, Score in Students.items():
        grade = CalculateGrade(Score)
        print(name, "\t", Score, "\t", grade)
    
```

## Update Student Score

This fucntion allows teachers to change an existing student's score

```python
def UpdateScore():
    name = input("Enter student name: ")

    if name not in Students:
        print("Student not found")
        return

    try:
        NewScore = float(input("Enter new score: "))
        if NewScore <0 or NewScore >100:
            print("Invalid score")
            return

        Students[name] = NewScore
        print("Score Updated!!")

    except ValueError:
        print("Please enter a valid number")
```

## Search Student

This function allows users to search for astudent's record.

```python
def SearchStudent():
    name = input("enter sudent name: ")
    if name in Students:
        Score = Students[name] 
        grade = CalculateGrade(Score)

        print("Student Found")
        print("Name:", name)
        print("Score:", Score)
        print("Grade:", grade)

    else:
        print("Student not Found!!")
```

## Class Statistics

This function calculates:
- Highest Score
- Lowest Score
- Average score
- Total number of Students

```python
def ClassStatistics():
    if not Students:
        print("No student records")
        return

    Scores = list(Students.values())

    Highest = max(Scores)
    Lowest = min(Scores)
    Average = sum(Scores) / len(Scores)

    print("Highest Score:", Highest)
    print("Lowest Score:", Lowest)
    print("Average Score:", round(Average, 2))
    print("Total Students:", len(Scores))
```

## Main Menu System
The menu keeps running until the user chooses "EXIT"

```python
def main():

    while True:

        print("\n==== STUDENT RESULT MANAGEMENT SYSTEM ====")
        print("1. Add Students")
        print("2. View Students")
        print("3. Update Score")
        print("4. Search Student")
        print("5. Class Statistics")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            AddStudent()
            
        elif choice == "2":
            ViewStudents()
            
        elif choice == "3":
            UpdateScore()

        elif choice == "4":
            SearchStudent()

        elif choice == "5":
            ClassStatistics()

        elif choice == "6":
            print("Goodbye!")
            break

        else:
            print("Invalid choice")
```

```python
#Dictionary to store products and quantities

inventory = {}
```

## Add Product Function

This function allows users to enter a product and quantity and store it in the inventory dictionary

```python
def add_product():
    
    product = input("Enter Product Name: ")

    try:
        quantity = int(input("Enter Quantity: "))

        if quantity < 0:
            print("Quantity cannot be negative!")
            return

        inventory[product] = quantity

        print("Product added Successfully!")

    except ValueError:
        print("Invalid Quantity Entered")
```

## View Inventory Function

This fuction displays all products and their quantities

```python
def view_inventory():
    
    if not inventory:
        print("Inventory is Empty / Product not Found.")
        return

    print("\nProduct\tQuantity")

    for product, quantity in inventory.items():
        
        print(product, "\t", quantity)
        
```

## Update Quantity Function

This function allows users to change the quantity of an existing product

```python
def update_quantity():
    
    product = input("Enter Product Name: ")
    
    if product not in inventory:
        print("Product not Found.")
        return

    try:
        
        new_quantity = int(input("Enter new quantity: "))

        if new_quantity < 0:
            print("Quantity cannot be negative.")
            return

        inventory[product] = new_quantity

        print("Quantity Updated Successfully!")

    except ValueError:
        print("Invalid Quantity entered.")
    
```

## Inventory Statistics Report Fuction

This function calculates inventory statistics.

```python
def inventory_stats_report():
    
    if not inventory:
        print("Inventory is Empty.")
        return

    quantities = list(inventory.values())

    total_products = len(inventory)

    total_stock = sum(quantities)

    highest_product = max(inventory, key=inventory.get)

    lowest_product = min(inventory, key=inventory.get)

    print("\nInventory Statistics")

    print("Total Products:", total_products)

    print("Product with Highest Quantity:", highest_product, "-", inventory[highest_product])

    print("Product with Lowest Quantity:", lowest_product, "-", inventory[lowest_product])

    print("Total Stock Available:", total_stock)
```
